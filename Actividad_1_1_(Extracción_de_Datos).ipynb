```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
#Cargar desde un archivo .xlsx sin indice
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

###Filtro 1

```python
#Filtrar todos los registros que en la columna NOMBRE_VENDEDOR tengan el
#nombre ALFREDO CANELA

filtro1  = df[df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA"]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv('Filtro1.csv')
```

```python
#descargar archivo filtrado en csv
from google.colab import files
files.download('Filtro1.csv')
```

###Filtro 2

```python
#Filtrar las filas desde el inidce 475 al 525

filtro2 = df.iloc[475:525,: ]
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv('Filtro2.csv')
```

```python
#descargar archivo filtrado en csv
from google.colab import files
files.download('Filtro2.csv')
```

###Filtro 3

```python
#Filtro para mostrar solo las columnas 2, 7 y 10

filtro3 = df.iloc[: , [2,7,10]]
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv('Filtro3.csv')
```

```python
#descargar archivo filtrado en csv
from google.colab import files
files.download('Filtro3.csv')
```

###Filtro 4

```python
#Paso 1: crear un nuevo dataframe con el nombre del vendedor como indice
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col = 2)
df1
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas
#Todas las filas con el indice DIEGO ARMANDO LEON DURAN de las columnas
#NOMBRE_CLIENTE, CANT y DESCR

filtro4 = df1.loc[['DIEGO ARMANDO LEON DURAN'], ['NOMBRE_CLIENTE', 'CANT', 'DESCR']]
filtro4
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv('Filtro4.csv')
```

```python
#descargar archivo filtrado en csv
from google.colab import files
files.download('Filtro4.csv')
```

###Filtro 6

```python
#Filtro por comparacion en donde el precio unitario sea mayor a 10000
filtro6 = df[df['PRECIO_UNITARIO'] > 10000]
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv('Filtro6.csv')
```

```python
#descargar archivo filtrado en csv
from google.colab import files
files.download('Filtro6.csv')
```

###Filtro 7

```python
#Filtro Y: Registros con precio unitario mayor a 8000
#y costo unitario mayor o igual a 2000
filtro7 = df[(df['PRECIO_UNITARIO'] > 8000) & (df['COSTO_UNITARIO'] >= 2000)]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv('Filtro7.csv')
```

```python
#descargar archivo filtrado en csv
from google.colab import files
files.download('Filtro7.csv')
```

###Filtro 8

```python
#Filtro O: registros que tengan o un margen unitario calculado mayor o igual a 15000
# o que el vendedor sea AARON NUÑO CORDOVA
filtro8 = df[(df['MARGEN_UNITARIO_CALCULADO'] >= 15000) | (df['NOMBRE_VENDEDOR'] == 'AARON NUÑO CORDOVA')]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv('Filtro8.csv')
```

```python
#descargar archivo filtrado en csv
from google.colab import files
files.download('Filtro8.csv')
```

###Filtro 9

```python
#Filtro not: Todos los registros con ventas de cantidades mayores a 50 unidades
# y que no sean de publico general
filtro9 = df[~(df['CANT'] <= 50) & ~(df['NOMBRE_CLIENTE'] == 'PÚBLICO EN GENERAL')]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv('Filtro9.csv')
```
